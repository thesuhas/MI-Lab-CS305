# Transform Data

```python
std = StandardScaler()
```

```python
std.fit(train)
```

# Getting Best SVM

```python
params = {'kernel': ['linear', 'rbf', 'poly', 'sigmoid'], 'C': [0.001, 0.05, 0.01, 0.1, 1, 5, 10, 100], 'gamma': [0.01, 0.1, 1, 5, 10, 100], 'degree': [1, 2, 3, 4, 5, 6]}
```

```python
svm = SVC()
```

```python
clf = GridSearchCV(svm, params, n_jobs = -1)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```
